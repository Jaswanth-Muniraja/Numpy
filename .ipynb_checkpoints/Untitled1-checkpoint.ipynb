### Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
print("=== Sales Analysis of a food delivery site ===")
print("\n sales data shape",sales_data.shape)
print("\n sample data for 1st 3 resturants", sales_data[0:3])
```

```python
print("\n sample data for 1st 3 resturants", sales_data[:,1:])
```

### total sales per year

```python
print(np.sum(sales_data,axis=0))
yearly_total = np.sum(sales_data[:,1:],axis = 0) # axis 0 is for column
print(yearly_total)
```

```python
print(np.sum(yearly_total))
```

```python
#### minimum sales per resturant
min_sales = np.min(sales_data[:,1:], axis = 1) # axis 1 is for row
print(min_sales)
```

```python
# max sales per year
max_sales = np.max(sales_data[:,1:], axis = 0) # axis 1 is for row
print(max_sales)
```

```python
# Average sale per restaurant
avg_sales = np.mean(sales_data[:,1:], axis = 1) # axis 1 is for row
print(avg_sales)
```

```python
# cumulative sum for restaurant
cumsum_sales = np.cumsum(sales_data[:,1:], axis = 1) # axis 1 is for row
print(cumsum_sales)
```

```python
print(np.mean(cumsum_sales,axis = 0))
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum_sales,axis = 0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print("Vector addition",vector1+vector2)
```

```python
print("Multiplication of vector",vector1*vector2)
```

```python
print("dot product of vector",np.dot(vector1,vector2))
```

```python
angle = np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(angle)
```
